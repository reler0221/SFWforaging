```python
import pandas as pd
import os
import ffmpeg
import glob
from pathlib import Path
import time
```

```python
#define the root folder where the search will start
root_folder = '/Volumes/T5 EVO/Foraging HD/Run_frame_extraction'

#define the folder where the frames will end up
dest_folder = '/Volumes/T5 EVO/Foraging HD/Raw_frames_to_annotate' #"Raw_frames_to_annotate" OR "Environment_frames_to_annotate"

#define the video file extensions you're interested in
video_extensions = ['.mp4', '.avi', '.mkv', '.mov', '.flv', '.wmv', '.MTS']

#list to store the paths of all video files
video_files = []

#recursively search for video files
for ext in video_extensions:
    video_files.extend(glob.glob(os.path.join(root_folder, '**', '*' + ext), recursive=True))

video_files
```

```python
#get all days
unique_dates=[f.split("/")[5] for f in video_files]
unique_dates=list(set(unique_dates))
unique_dates
```

```python
#to count pictures in folder

def count_jpg_files(folder_path):
    return len(list(Path(folder_path).glob('*.jpg')))
```

```python
#batch starting name
index_name=88
i=0
time_start=time.time()
for i in range(0,len(video_files)):
    print(str(i)+" out of "+ str(len(video_files)))
    image_name="_".join(video_files[i].split("/")[6:]).replace(".MTS","").replace(".","_")+'_%04d.jpg'
    #image_name

    imagedir=f"{dest_folder}"+"/Batch"+str(index_name)
    #imagedir

    if not os.path.exists(imagedir):
        os.makedirs(imagedir)

    output_pattern_out=imagedir+"/"+image_name

    (
        ffmpeg
        .input(video_files[i], ss=0)#, t=3)
        .filter('yadif', 0)
        .filter('fps', fps=2)#change here to collect more or less data
        .output(output_pattern_out)
        .run()
    )

    #if more than 1000 pictures alread stored in a single folder change to another
    if count_jpg_files(imagedir)>1000:
        index_name=index_name+1
time_end=time.time()
extraction_duration=(time_end-time_start)/60 #duration of the run in minutes
print("Done",extraction_duration)
        
```
